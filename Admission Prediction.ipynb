```python
def hasAdmitted(data):
    if data > 0.7:
        return 1
    else:
        return 0
admission['Admit'] = admission['Chance of Admit'].apply(hasAdmitted)
admission.head()
```

```python
X=admission.drop(['Admit'],axis=1)
y=admission['Admit']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,
                                               test_size=0.3,random_state=70)
```

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
y_pred=lr.predict(X_test)
```

```python
y_test
```

```python
y_pred
```

```python
lr.score(X_test,y_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,y_pred)
```

# Linear Regression

```python
x=admission[['CGPA']]
```

```python
y=admission[['GRE Score']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,
                                               test_size=0.2,random_state=50)
```

```python
x_train.size
```

```python
y_train.size
```

```python
x_test.size
```

```python
y_test.size
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression.fit(x_train,y_train)
```

```python
y_pred=regression.predict(x_test)
```

```python
y_test
```

```python
y_pred
```

```python
plt.scatter(x_train,y_train,c='blue')
plt.plot(x_train,regression.predict(x_train),c='red')
plt.title("CGPAvsGREScore(Training)")
plt.xlabel("CGPA")
plt.ylabel("GRE Score")
```

```python
plt.scatter(x_test,y_test,c='blue')
plt.plot(x_train,regression.predict(x_train),c='red')
plt.title("CGPAvsGREScore(Training)")
plt.xlabel("CGPA")
plt.ylabel("GRE Score")
```

```python
from sklearn import metrics
rmse=np.sqrt(metrics.mean_squared_error(y_test,y_pred))
```

```python
rmse
```

```python
R2=metrics.r2_score(y_test,y_pred)
```

```python
R2
```

# Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
dt=DecisionTreeRegressor()
```

```python
dt.fit(x_train,y_train)
```
